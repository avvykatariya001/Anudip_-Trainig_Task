```python
# Question 1
plt.figure(figsize=(12, 5))
dataset.groupby('year')['amount'].count().plot(kind='bar', title='Year Wise Sales', color='Red', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question 2
dataset_2015_2018 = dataset[(dataset['year'] >= 2015) & (dataset['year'] <= 2018)]

plt.figure(figsize=(10, 5))
dataset_2015_2018.groupby('month')['rating'].count().plot(kind='bar', color='teal', edgecolor='black')
plt.title('Month Wise Sales (2015–2018)')
plt.xlabel('Month')
plt.ylabel('Number of Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Question 3
dataset_2015_2018 = dataset[(dataset['year'] >= 2015) & (dataset['year'] <= 2018)]

plt.figure(figsize=(12, 5))
dataset_2015_2018.groupby('brand')['amount'].sum().sort_values(ascending=False).head(10)    .plot(kind='bar', title='Brand Wise Top 10 Sales 2015 to 2018', color='green', edgecolor='black')
plt.xlabel('Brand')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question 4
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

top_selling_2016 = dataset[dataset['year'] == 2016].groupby('brand')['rating'].count().sort_values(ascending=False).head(10)
axs[0, 0].bar(top_selling_2016.index, top_selling_2016, color='steelblue', edgecolor='black')
axs[0, 0].set_title('Top Selling Brands in 2016')
axs[0, 0].tick_params(axis='x', rotation=45)

top_selling_2017 = dataset[dataset['year'] == 2017].groupby('brand')['rating'].count().sort_values(ascending=False).head(10)
axs[0, 1].bar(top_selling_2017.index, top_selling_2017, color='darkorange', edgecolor='black')
axs[0, 1].set_title('Top Selling Brands in 2017')
axs[0, 1].tick_params(axis='x', rotation=45)

top_selling_2018 = dataset[dataset['year'] == 2018].groupby('brand')['rating'].count().sort_values(ascending=False).head(10)
axs[1, 0].bar(top_selling_2018.index, top_selling_2018, color='green', edgecolor='black')
axs[1, 0].set_title('Top Selling Brands in 2018')
axs[1, 0].tick_params(axis='x', rotation=45)

axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Question5
dataset2015_2018 = dataset[(dataset['year'] >= 2015) & (dataset['year'] <= 2018)]

plt.figure(figsize=(12, 5))
dataset2015_2018.groupby('category')['amount'].sum().sort_values(ascending=False).head(10)    .plot(kind='bar', title='Top 10 Most Sold Product Category 2015 to 2018', color='mediumseagreen', edgecolor='black')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question 9
dataset2015_2018 = dataset[(dataset['year'] >= 2015) & (dataset['year'] <= 2018)]

plt.figure(figsize=(12, 5))
dataset2015_2018.groupby('brand')['rating'].mean().sort_values(ascending=False).head(10)    .plot(kind='bar', title='10 Most Highly Rated Brands 2015 to 2018', color='mediumpurple', edgecolor='black')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.ylim(0, 5.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Quesiton 10
plt.figure(figsize=(8, 8))
dataset.groupby('category')['amount'].sum()
.sort_values(ascending=False).head(5)    
.plot(kind='pie', autopct='%1.1f%%', title='Top 5 Category Sales Percentage', startangle=90)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Question 11
plt.figure(figsize=(8, 8))
dataset.groupby('brand')['rating'].count().sort_values(ascending=False).head(5)    .plot(kind='pie', autopct='%1.1f%%', title='Top 5 Brand Wise Sales Percentage', startangle=90)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Question 12
gender_distribution = dataset['gender'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(gender_distribution, labels=gender_distribution.index,
        autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Gender Wise Customer Distribution')
plt.show()
```
